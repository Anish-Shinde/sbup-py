***LIST***

```python
marks = [76,45,76,22,45]
print(marks)
print(type(marks))
print(marks[0])
print(marks[3])
print(len(marks))
```

```python
students = ["Sanket",98,23,"Pune"]
print(students)
```

```python
#List is Mutable they can change anything
students = ["Sanket",98,23,"Pune"]
print(students[0])
students[0] = "Banket"
print(students)
```

```python
#In List slicing is allowed
marks = [76,45,76,22,45]
print(marks[1:3])
```

```python
#list.append()....adds one elements at the end..yaha element add kr skte hai list mai
marks = [76,45,76,22,45]
marks.append(34)
print(marks)
```

```python
#list.sort()...sorts in ascending order...
marks = [76,45,76,22,45]
marks.sort()
print(marks)
```

```python
# list.sort(reverse= True)...sorts in decending order
marks = [76,45,76,22,45]
marks.sort(reverse = True)
print(marks)
```

```python
#list.reverse()...reverse the list
marks = [76,45,76,22,45]
marks.reverse()
print(marks)
```

```python
#list.insert(indix,element)...insert element at index
marks = [76,45,70,22,45]
marks.insert(1,77)
print(marks)
```

```python
#list.remove()...remove first occurrence of element
marks = [76,45,76,22,45]
marks.remove(45)
print(marks)
```

```python
#list.pop()....remove element at index
marks = [76,45,76,22,45]
marks.pop(2)
print(marks)
```

```python
#NESTED DIC..
student = {
    "name": "Sanket",
    "Score": {
        "math": 98,
        "chem": 85,
        "bio": 97,
    }
}
print(student)
print(student["Score"])
print(type(student))
```

```python
#dic.keys()...return all keys
student = {
    "name": "Sanket",
    "Score": {
        "math": 98,
        "chem": 85,
        "bio": 97,
    }
}
print(student.keys())
```

```python
#dic.values()...return all values
student = {
    "name": "Sanket",
    "Score": {
        "math": 98,
        "chem": 85,
        "bio": 97,
    }
}
print(student.values())
```

```python
#dic.items()...return aal key value pairs as tuples
student = {
    "name": "Sanket",
    "Score": {
        "math": 98,
        "chem": 85,
        "bio": 97,
    }
}
print(student.items())
```

```python
#dic.update(new dic)...inserts the specified items to the dic
student = {
    "name": "Sanket",
    "Score": {
        "math": 98,
        "chem": 85,
        "bio": 97,
    }
}
student.update({"city": "Pune"})
print(student)

```

***SETS***
set is unordered items
elements is unique and immutables
set is mutable but their elements are immutables

```python
collection = {1,2,3,4}
print(collection)
print(type(collection))
```

```python
#in set ..set are remove the duplicate values
collection = {1,2,3,4,"sk","sk","sk",2,2,3}
print(collection)
```

```python
#set.add(element)...adds an elements

collection = set()
collection.add(1)
collection.add(2)
collection.add("Sanket")
collection.add((1,2,3,4))

print(collection)
```

```python
#set.remove(el)..removes the element

collection = {1,2,3,4}
collection.remove(3)
print(collection)
print(len(collection))
```

```python
#set.clear()...remove all set
collection = set()
collection.add(1)
collection.add(2)
collection.add("Sanket")
collection.add((1,2,3,4))

collection.clear()
print(collection)
```

```python
#set.pop()...remove random value
collection ={"hello","sanket","how","are","you"}
print(collection.pop())
print(collection.pop())
```

```python
#set.union(set2)...combines both set values & return new 
set1 = {1,2,3}
set2 = {2,3,4}
print(set1.union(set2))
```

```python
#set.intersection(set2)...combine common values and return new
set1 = {1,2,3}
set2 = {2,3,4}
print(set1.intersection(set2))
```
